# Statistics Basics

1. What is the difference between descriptive statistics and inferential statistics? Explain with examples.
   - Descriptive statistics describe and summarize data we already have — like
     finding the average marks or percentage of students who passed in a class.

   - Inferential statistics use a sample to make predictions or conclusions
     about a larger group — like predicting the average marks of all students in a school based on one class.

2. What is sampling in statistics? Explain the differences between random and
   stratified sampling.
   - Sampling in statistics means selecting a small group (sample) from a large
     population to study, because studying everyone is often not possible.
   - In random sampling, every member of the population has an equal chance of
     being selected.
   - In stratified sampling, the population is divided into groups (strata)
     based on some characteristics (like gender or age), and then samples are taken from each group.

3. Define mean, median, and mode. Explain why these measures of central
   tendency are important.
   - Mean is the average of all numbers in a data set.
     - Example: The mean of 10, 20, and 30 is 20.

   - Median is the middle value when the numbers are arranged in order.
     - Example: The median of 10, 20, and 30 is 20.

   - Mode is the number that appears most often.
     - Example: The mode of 10, 20, 20, and 30 is 20.

   - These are important because they show the center or typical value of the
     data and help us understand it easily.

4. Explain skewness and kurtosis. What does a positive skew imply about the
   data?
   - Skewness tells us how the data is spread around the mean — whether it
     leans more to one side.
     - Positive skew means the tail is longer on the right side. Most values are
       small, and a few large values pull the mean to the right.
     - Negative skew means the tail is longer on the left side. Most values are
       high, but a few small ones pull the mean to the left.
   
   - Kurtosis tells us how peaked or flat a distribution is compared to a
     normal distribution.
     - High kurtosis = more peak, more outliers.
     - Low kurtosis = flatter curve, fewer outliers.

      

```python
#5	Implement a Python program to compute the mean, median, and mode of a given list of numbers.
#   numbers = [12, 15, 12, 18, 19, 12, 20, 22, 19, 19, 24, 24, 24, 26, 28]

import statistics as stats

numbers = [12, 15, 12, 18, 19, 12, 20, 22, 19, 19, 24, 24, 24, 26, 28]

mean = stats.mean(numbers)
median = stats.median(numbers)
mode = stats.mode(numbers)

print("Mean:", mean)
print("Median:", median)
print("Mode:", mode)

```

```python
#6 Compute the covariance and correlation coefficient between the following two datasets provided as lists in Python:
#  list_x = [10, 20, 30, 40, 50]
#  list_y = [15, 25, 35, 45, 60]

import numpy as np

list_x = [10, 20, 30, 40, 50]
list_y = [15, 25, 35, 45, 60]

covariance = np.cov(list_x, list_y)[0, 1]
correlation = np.corrcoef(list_x, list_y)[0, 1]

print("Covariance:", covariance)
print("Correlation Coefficient:", correlation)

```

```python
#7 Write a Python script to draw a boxplot for the following numeric list and identify its outliers. Explain the result:
#  data = [12, 14, 14, 15, 18, 19, 19, 21, 22, 22, 23, 23, 24, 26, 29, 35]

import matplotlib.pyplot as plt
import numpy as np

data = [12, 14, 14, 15, 18, 19, 19, 21, 22, 22, 23, 23, 24, 26, 29, 35]

plt.boxplot(data)
plt.title("Boxplot of Data")
plt.ylabel("Values")
plt.show()

Q1 = np.percentile(data, 25)
Q3 = np.percentile(data, 75)
IQR = Q3 - Q1

lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

outliers = [x for x in data if x < lower_bound or x > upper_bound]

print("Outliers:", outliers)

```

The boxplot shows the spread of the data with the median and quartiles.

Using the Interquartile Range (IQR) rule, any value greater than Q3 + 1.5 × IQR or less than Q1 - 1.5 × IQR is an outlier.

Here, 35 lies above the upper limit, so it is an outlier.

```python
#8 You are working as a data analyst in an e-commerce company. The marketing team wants to know if there is a relationship between advertising spend and daily sales.
#  ● Explain how you would use covariance and correlation to explore this relationship.
#  ● Write Python code to compute the correlation between the two lists:
#  advertising_spend = [200, 250, 300, 400, 500]
#  daily_sales = [2200, 2450, 2750, 3200, 4000]
```

- Covariance
  - It shows how two variables change together.
  - If covariance > 0, both increase together (positive relationship).
  - If covariance < 0, one increases while the other decreases
     (negative relationship).
   - But covariance doesn’t tell how strong the relationship is.

- Correlation
  - It measures the strength and direction of the relationship between two
     variables.
  - Value lies between -1 and +1:
  - +1 → strong positive relationship
  - -1 → strong negative relationship
  - 0 → no relationship

```python
import numpy as np

advertising_spend = [200, 250, 300, 400, 500]
daily_sales = [2200, 2450, 2750, 3200, 4000]

covariance = np.cov(advertising_spend, daily_sales)[0, 1]

correlation = np.corrcoef(advertising_spend, daily_sales)[0, 1]

print("Covariance:", covariance)
print("Correlation Coefficient:", correlation)
```

```python
#9 Your team has collected customer satisfaction survey data on a scale of 1-10 and wants to understand its distribution before launching a new product.
#  ● Explain which summary statistics and visualizations (e.g. mean, standard deviation, histogram) you’d use.
#  ● Write Python code to create a histogram using Matplotlib for the survey data:
#    survey_scores = [7, 8, 5, 9, 6, 7, 8, 9, 10, 4, 7, 6, 9, 8, 7]
```

- Summary Statistics
  - Mean - shows the average satisfaction level.
  - Median - shows the middle value, useful if there are extreme scores.
  - Mode - shows the most common rating given by customers.
  - Standard Deviation - shows how spread out the scores are
     (small value = most customers gave similar ratings).
  - Minimum and Maximum - show the range of satisfaction levels.

- Visualizations
  - Histogram - shows the frequency distribution of scores, helping to see if
     ratings are high, low, or evenly spread.
  - Boxplot - shows the spread, median, and outliers clearly.

```python
import matplotlib.pyplot as plt
import numpy as np
from statistics import mean, median, mode, stdev

survey_scores = [7, 8, 5, 9, 6, 7, 8, 9, 10, 4, 7, 6, 9, 8, 7]

print("Mean:", mean(survey_scores))
print("Median:", median(survey_scores))
print("Mode:", mode(survey_scores))
print("Standard Deviation:", round(stdev(survey_scores), 2))
print("Min:", min(survey_scores))
print("Max:", max(survey_scores))


plt.hist(survey_scores, bins=6, color='skyblue', edgecolor='black')
plt.title("Customer Satisfaction Score Distribution")
plt.xlabel("Survey Scores (1–10)")
plt.ylabel("Frequency")
plt.show()
```
