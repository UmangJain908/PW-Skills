15. What is causation? Explain difference between correlation and causation
     with an example.
    - Causation: One variable directly affects another.
    - Correlation: Variables move together, but one does not necessarily cause
      the other.
    - Example:
      - Umbrella sales increase when rainfall increases.
      - This is correlation, but umbrellas do not cause rain.
      - The true cause behind both is weather conditions (rainy weather).

16. What is an Optimizer? What are different types of optimizers? Explain each  
     with an example.
    - An optimizer is a method/algorithm that adjusts the model’s parameters  
      (weights) during training to reduce the loss and make predictions better.
    - Example:
      - Gradient Descent (GD)
        - Updates weights in the direction where loss decreases.
        - Example: Basic optimizer used in linear regression.
      - Stochastic Gradient Descent (SGD)
        - Updates weights using one sample at a time -> faster but noisier.
        - Example: Used in simple neural networks.
      - Mini-Batch Gradient Descent
        - Updates weights using small batches of data -> balances speed and accuracy.
        - Example: Most deep learning models use this.
      - Adam (Adaptive Moment Estimation)
        - Fast, efficient optimizer that adjusts the learning rate
          automatically.
        - Example: Most commonly used in deep learning (CNNs, RNNs).
