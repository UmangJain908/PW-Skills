# Data Toolkit Questions

1. What is NumPy, and why is it widely used in Python?
   - NumPy (Numerical Python) is a library used for numerical computations.  
   - It provides multi-dimensional arrays and mathematical functions.  
   - It’s fast, memory-efficient, and integrates well with other libraries like
     Pandas and Scikit-learn.  

2. How does broadcasting work in NumPy?
   - Broadcasting allows arithmetic operations between arrays of different
     shapes.  
   - Smaller arrays are automatically expanded to match the shape of larger
     arrays.  

3. What is a Pandas DataFrame?  
   - A DataFrame is a two-dimensional labeled data structure with rows and
     columns.  
   - It’s used for storing, manipulating, and analyzing tabular data.  

4. Explain the use of the 'groupby()' method in Pandas.
   - The 'groupby()' method splits data into groups and applies functions (like
     sum, mean, count) on each group.  
   - Example: 'df.groupby('Category')['Sales'].sum()'  

5. Why is Seaborn preferred for statistical visualizations?
   - Seaborn simplifies creating complex plots and integrates with Pandas.  
   - It offers attractive default styles and specialized plots for data
     distributions.  

6. What are the differences between NumPy arrays and Python lists?
   - NumPy arrays are faster and more memory-efficient.  
   -  Arrays support element-wise operations and broadcasting.  
   - Lists can store mixed data types, arrays cannot.  

7. What is a heatmap, and when should it be used?  
   - A heatmap visualizes data using color gradients.  
   - It’s useful for showing correlations or matrix-like data (e.g., confusion
     matrix).  

8. What does the term “vectorized operation” mean in NumPy?
   - Performing operations on entire arrays without using loops.  
   - It speeds up execution and simplifies code.  

9. How does Matplotlib differ from Plotly?
   - Matplotlib creates static, publication-style plots.  
   - Plotly creates interactive, web-based visualizations with hover and zoom.  

10. What is the significance of hierarchical indexing in Pandas?
    - Hierarchical indexing (MultiIndex) allows multiple levels of index.  
    - It’s useful for representing higher-dimensional data compactly.  

11. What is the role of Seaborn’s 'pairplot()' function?
    - It visualizes pairwise relationships among numerical variables.  
    - Useful for identifying correlations and trends.  

12. What is the purpose of the 'describe()' function in Pandas?  
    - Provides a quick statistical summary of numeric columns.  
    - Includes count, mean, std, min, and quartile values.  

13. Why is handling missing data important in Pandas?
    - Missing data can distort analysis and model accuracy.  
    - Functions like 'isna()', 'fillna()', and 'dropna()' help handle it.  

14. What are the benefits of using Plotly for data visualization?
    - Interactive and dynamic visualizations.  
    - Works in browsers and dashboards.  
    - Supports 3D plots and animations.  

15. How does NumPy handle multidimensional arrays?
    - NumPy uses 'ndarray' objects for N-dimensional arrays.  
    - It supports slicing, reshaping, and mathematical operations across
      dimensions.  

16. What is the role of Bokeh in data visualization?  
    - Bokeh creates interactive, web-based, and real-time visualizations.  
    - Ideal for dashboards and live data applications.  

17. Explain the difference between 'apply()' and 'map()' in Pandas.  
    - 'map()' works on Series and applies a function element-wise.  
    - 'apply()' works on both Series and DataFrames for row/column-wise
      transformations.  

18. What are some advanced features of NumPy?  
    - Linear algebra, random number generation, and Fourier transforms.  
    - Broadcasting and memory-efficient views.  
    - Integration with C/C++ for high-performance computing.  

19. How does Pandas simplify time series analysis?
    - Built-in support for date-time indexing and resampling.  
    - Provides rolling, shifting, and trend analysis functions.  

20. What is the role of a pivot table in Pandas?
    - Summarizes data similar to Excel pivot tables.  
    - Aggregates data using specified keys and functions.  

21. Why is NumPy’s array slicing faster than Python’s list slicing?  
    - NumPy arrays store data in contiguous memory blocks.  
    - Slicing creates a view instead of a copy, improving speed and
      efficiency.  

22. What are some common use cases for Seaborn?  
    - Visualizing distributions ('histplot', 'kdeplot').  
    - Pairwise relationships ('pairplot').  
    - Correlations ('heatmap').  
    - Categorical comparisons ('boxplot', 'violinplot').  
    - Time series and regression visualizations.  


# Practical

```python
#1 How do you create a 2D NumPy array and calculate the sum of each row?

import numpy as np

arr = np.array([[1, 2, 3],
                [4, 5, 6]])

row_sum = np.sum(arr, axis=1)
print(row_sum)
```

```python
#2 Write a Pandas script to find the mean of a specific column in a DataFrame?

import pandas as pd

data = {'A': [1, 2, 3],
        'B': [4, 5, 6]}

df = pd.DataFrame(data)

mean_B = df['B'].mean()
print(mean_B)
```

```python
#3 Create a scatter plot using Matplotlib?

import matplotlib.pyplot as plt

x = [1, 2, 3, 4, 5]
y = [2, 4, 6, 8, 10]

plt.scatter(x, y)
```

```python
#4 How do you calculate the correlation matrix using Seaborn and visualize it with a heatmap?

import seaborn as sns

data = {'A': [10, 20, 30],
        'B': [44, 55, 66]}

df = pd.DataFrame(data)

correlation_matrix = df.corr()

sns.heatmap(correlation_matrix, annot=True)
plt.show()
```

```python
#5 Generate a bar plot using Plotly?

import plotly.graph_objects as go

x = ['A', 'B', 'C']
y = [10, 20, 30]

fig = go.Figure(data=[go.Bar(x=x, y=y)])

fig.show()
```

```python
#6 Create a DataFrame and add a new column based on an existing column.

data = {'A': [1, 2, 3],
        'B': [4, 5, 6]}

df = pd.DataFrame(data)

df['C'] = df['A'] + df['B']
print(df)
```

```python
#7 Write a program to perform element-wise multiplication of two NumPy arrays.

arr1 = np.array([1, 2, 3, 4])
arr2 = np.array([5, 6, 7, 8])
result = arr1 * arr2
print(result)
```

```python
#8 Create a line plot with multiple lines using Matplotlib.

x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.cos(x)

plt.plot(x, y1, label='sin(x)')
plt.plot(x, y2, label='cos(x)')
plt.legend()
plt.show()
```

```python
#9 Generate a Pandas DataFrame and filter rows where a column value is greater than a threshold.

data = {'A': [1, 2, 3, 4, 5],
        'B': [10, 20, 30, 40, 50]}

df = pd.DataFrame(data)

threshold = 25
filtered_df = df[df['B'] > threshold]
print(filtered_df)

```

```python
#10 Create a histogram using Seaborn to visualize a distribution.

import seaborn as sns
import matplotlib.pyplot as plt

data = [1, 2, 2, 3, 3, 3, 4, 4, 5]

sns.histplot(data, bins=5)
plt.show()
```

```python
#11 Perform matrix multiplication using NumPy?

A = np.array([[1, 2],
              [3, 4]])

B = np.array([[5, 6],
              [7, 8]])

result = np.dot(A, B)

print(result)
```

```python
#13 Create a 3D scatter plot using Plotly.

import plotly.graph_objects as go

x = [1, 2, 3, 4, 5]
y = [10, 15, 13, 17, 12]
z = [5, 8, 6, 9, 7]

fig = go.Figure(data=[go.Scatter3d(
    x=x,
    y=y,
    z=z,
    mode='markers')])

fig.show()
```
