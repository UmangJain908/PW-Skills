# Supervised Learning: Regression Models and Performance Metrics.

1. What is Simple Linear Regression (SLR)? Explain its purpose.
   - Simple Linear Regression (SLR) is a supervised learning method used to predict a value using one input variable.
   - It finds a straight-line relationship between an independent variable (X) and a dependent variable (Y).
   - Purpose:
      To understand the relationship between two variables
      To predict future values
   - Formula: Y=mX+c
   - Example: Predict marks from study hours.

2. What are the key assumptions of Simple Linear Regression?
   - Key assumptions of Simple Linear Regression (SLR):
   - Linear relationship – X and Y have a straight-line relationship.
   - Independence – Observations are independent of each other.
   - Homoscedasticity – Constant variance of errors.
   - Normality of errors – Residuals are normally distributed.
   - No outliers – Extreme values should not strongly affect the model.

3. Write the mathematical equation for a simple linear regression model and
   explain each term.
    - Simple Linear Regression Equation: Y=mX+c
    - Explanation of terms:
      - Y – Dependent variable (output to be predicted)
      - X – Independent variable (input)
      - m – Slope (change in Y for one unit change in X)
      - c – Intercept (value of Y when X = 0)

4. Provide a real-world example where simple linear regression can be
   applied?
    - Real-world example of Simple Linear Regression:
      - Predicting salary based on years of experience.
      - X (independent variable): Years of experience
      - Y (dependent variable): Salary


5. What is the method of least squares in linear regression?
    - The method of least squares is used to find the best-fit line in linear
       regression.
   - It works by minimizing the sum of squared differences between the actual
       values and the predicted values.

6. What is Logistic Regression? How does it differ from Linear Regression?
    - Logistic Regression is a supervised learning algorithm used for classification, especially for binary outcomes like yes/no or pass/fail. It predicts the probability of an event using a sigmoid (S-shaped) function.
    - Linear Regression, on the other hand, is used for predicting continuous values such as salary or marks and uses a straight-line equation.
    - The main difference is that Linear Regression gives numerical outputs, while Logistic Regression gives class-based outputs (0 or 1)

7. Name and briefly describe three common evaluation metrics for regression
   models.
    - Three common evaluation metrics for regression models are Mean Absolute Error (MAE), Mean Squared Error (MSE), and R-squared (R²). MAE measures the average absolute difference between actual and predicted values. MSE calculates the average of squared errors, giving more weight to large errors. R-squared shows how well the model explains the variability in the data, with values closer to 1 indicating a better fit.

8. What is the purpose of the R-squared metric in regression analysis?
   - The purpose of the R-squared (R²) metric is to measure how well a regression model explains the variation in the dependent variable. It shows the proportion of data variance explained by the model. An R-squared value closer to 1 means a better model fit, while a value closer to 0 means a poor fit.


```python
#9 Write Python code to fit a simple linear regression model using scikit-learn and print the slope and intercept.

import numpy as np
from sklearn.linear_model import LinearRegression

# Sample data
X = np.array([1, 2, 3, 4, 5]).reshape(-1, 1)
y = np.array([2, 4, 6, 8, 10])

# Create and train model
model = LinearRegression()
model.fit(X, y)

# Print slope and intercept
print("Slope (m):", model.coef_[0])
print("Intercept (c):", model.intercept_)

```

10.  How do you interpret the coefficients in a simple linear regression model?
     - In a simple linear regression model, the coefficients tell us how the dependent variable (Y) changes with the independent variable (X):
     - Slope (m): Shows how much Y changes for a one-unit increase in X.
       - Example: If slope = 2, then Y increases by 2 units for every 1-unit increase in X.

     - Intercept (c): Shows the value of Y when X = 0.
       - Example: If intercept = 5, then when X is 0, Y is 5.
