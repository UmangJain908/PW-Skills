1. What is hypothesis testing in statistics?
   - Hypothesis testing is a statistical method used to make decisions about a
     population based on sample data. We start with an assumption called the null hypothesis and check if the sample provides enough evidence to reject it.

2. What is the null hypothesis, and how does it differ from the alternative    
   hypothesis?
   - The null hypothesis (H₀) assumes there is no effect, no difference, or no
     relationship in the data. The alternative hypothesis (H₁) claims the opposite—that an effect or difference exists. Hypothesis testing checks whether sample evidence is strong enough to reject H₀ and support H₁, helping us make decisions statistically.

3. Explain the significance level in hypothesis testing and its role in deciding
   the outcome of a test.
   - The significance level (α) is the probability of rejecting the null
     hypothesis when it is actually true. It acts as a threshold (commonly 0.05) to decide whether results are statistically meaningful. If the p-value is less than α, we reject the null hypothesis, indicating the result is unlikely due to chance.

4. What are Type I and Type II errors? Give examples of each.
   - Type I error happens when we reject a true null hypothesis.
     - Example: concluding a new medicine works when it actually doesn’t.
   - Type II error happens when we fail to reject a false null hypothesis.
     - Example: saying a medicine doesn’t work when it actually does.
  
5. What is the difference between a Z-test and a T-test? Explain when to use
   each.
   - A Z-test is used when the sample size is large (n > 30) and the population
     standard deviation is known.
   - A T-test is used when the sample size is small (n ≤ 30) and the population
     standard deviation is unknown.

```python
#7 Implement hypothesis testing using Z-statistics for a sample dataset in Python. Show the Python code and interpret the results.
#  sample_data = [49.1, 50.2, 51.0, 48.7, 50.5, 49.8, 50.3, 50.7, 50.2, 49.6, 50.1, 49.9, 50.8, 50.4, 48.9, 50.6, 50.0, 49.7, 50.2, 49.5,50.1, 50.3, 50.4, 50.5,
#                 50.0, 50.7, 49.3, 49.8, 50.2, 50.9, 50.3, 50.4, 50.0, 49.7, 50.5, 49.9]

import numpy as np
from scipy.stats import norm

sample_data = [49.1, 50.2, 51.0, 48.7, 50.5, 49.8, 50.3, 50.7, 50.2, 49.6,
               50.1, 49.9, 50.8, 50.4, 48.9, 50.6, 50.0, 49.7, 50.2, 49.5,
               50.1, 50.3, 50.4, 50.5, 50.0, 50.7, 49.3, 49.8, 50.2, 50.9,
               50.3, 50.4, 50.0, 49.7, 50.5, 49.9]

mu_0 = 50

sample_mean = np.mean(sample_data)
sample_std = np.std(sample_data, ddof=1)
n = len(sample_data)

z_stat = (sample_mean - mu_0) / (sample_std / np.sqrt(n))

p_value = 2 * (1 - norm.cdf(abs(z_stat)))

print("Sample Mean:", round(float(sample_mean), 4))
print("Sample Std Dev:", round(float(sample_std), 4))
print("Z-Statistic:", round(float(z_stat), 4))
print("P-Value:", round(float(p_value), 4))
```

```python
#8 Write a Python script to simulate data from a normal distribution and calculate the 95% confidence interval for its mean. Plot the data using Matplotlib.

import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

np.random.seed(42)
data = np.random.normal(loc=50, scale=5, size=500)

mean = np.mean(data)
std = np.std(data, ddof=1)
n = len(data)

z = 1.96
lower = mean - z * (std / np.sqrt(n))
upper = mean + z * (std / np.sqrt(n))

(mean, lower, upper)

plt.hist(data, bins=30)
plt.title("Normal Distribution Data")
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.show()
```

```python
#9 Write a Python function to calculate the Z-scores from a dataset and visualize the standardized data using a histogram. Explain what the Z-scores represent
#  in terms of standard deviations from the mean.

import numpy as np
import matplotlib.pyplot as plt

def calculate_z_scores(data):
    mean = np.mean(data)
    std = np.std(data, ddof=1)
    z_scores = (data - mean) / std
    return z_scores

np.random.seed(42)
data = np.random.normal(50, 5, 300)

z_scores = calculate_z_scores(data)

plt.hist(z_scores, bins=30)
plt.title("Histogram of Z-Scores")
plt.xlabel("Z-Score")
plt.ylabel("Frequency")
plt.show()
```

Z-scores tell how far a data point is from the mean, measured in standard deviations.

Z = 0 → exactly at mean
Z = +1 → one standard deviation above mean
Z = -2 → two standard deviations below mean
Z-scores help standardize data so different scales can be compared easily.
