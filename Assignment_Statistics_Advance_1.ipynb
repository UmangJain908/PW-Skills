1. What is a random variable in probability theory?
   - A random variable is a variable whose value is not fixed ‚Äî it changes
     based on the outcome of a random experiment.
     - Example: When you roll a dice, the random variable X can take values 1
       to 6 depending on what you get.

2. What are the types of random variables?
   - There are two main types of random variables:
     - Discrete Random Variable
       - Takes countable values.
       - Example: Dice outcomes (1, 2, 3, 4, 5, 6).
     - Continuous Random Variable
       - Takes uncountable values within a range.
       - Example: Height, temperature, time.

3. Explain the difference between discrete and continuous distributions.
   - Discrete Distribution
     - Deals with countable outcomes.
     - You can list the values.
     - Example: Number of heads in 5 coin tosses.

   - Continuous Distribution
     - Deals with uncountable values in a range.
     - Values come from a continuous interval.
     - Example: Time taken to run 1 km.

4. What is a binomial distribution, and how is it used in probability?
   - A binomial distribution is a probability distribution that tells us the
     chance of getting a certain number of successes in a fixed number of independent trials, where each trial has only two outcomes (success or failure).

   - Conditions:
     - Fixed number of trials (n)
     - Each trial has 2 outcomes (success/failure)
     - Probability of success (p) stays the same
     - Trials are independent

     - Example:
       - Flipping a coin 10 times and counting how many heads you get.

    - Use:
      - It helps calculate the probability of getting k successes out of n
         trials, like:
      - Number of defective items in a batch
      - Number of students passing an exam
      - Number of goals scored in fixed attempts

5. What is the standard normal distribution, and why is it important?
   - The standard normal distribution is a normal (bell-shaped) distribution
     with:
     - Mean = 0
     - Standard deviation = 1
   - Importance of the standard normal distribution:

     - Easy probability calculations using Z-tables.
     - Standard way to compare different datasets by converting values to
       Z-scores.
     - Basis of many statistical tests like Z-test and confidence intervals.
     - Many real-world measurements (like heights, errors) follow or
       approximate a normal curve.

6. What is the Central Limit Theorem (CLT), and why is it critical in
   statistics?
   - It says that when you take many samples from any population and calculate
     their means, those sample means will form a normal (bell-shaped) distribution, even if the original data is not normal.

   - Importance:
     - It lets us use normal distribution formulas for many problems.
     - It helps in hypothesis testing and confidence intervals.
     - It makes statistics work even when the data is messy or not normal.

7. What is the significance of confidence intervals in statistical analysis?
   - Confidence intervals give a range of values that likely contains the true
     population value.

   - Significance:
     - Gives uncertainty clearly ‚Äî not just one estimate, but a range.
     - Shows reliability of an estimate (narrow interval = more accurate).
     - Used in decision-making ‚Äî helps judge if results are meaningful.
     - Works with sample data when the whole population can‚Äôt be measured.

8. What is the concept of expected value in a probability distribution.
   - The expected value is the average outcome you would get if you repeated an
     experiment many times.
   - Example: For a fair dice:
     - E(X)=(1+2+3+4+5+6)/6=3.5
     - So the expected value = 3.5, even though you can‚Äôt roll 3.5.


```python
#9 Write a Python program to generate 1000 random numbers from a normal distribution with mean = 50 and standard deviation = 5.
#  Compute its mean and standard deviation using NumPy, and draw a histogram to visualize the distribution.

import numpy as np
import matplotlib.pyplot as plt

data = np.random.normal(50, 5, 1000)

print("Mean:", np.mean(data))
print("Standard Deviation:", np.std(data))

plt.hist(data, bins=30)
plt.show()
```

```python
#10 You are working as a data analyst for a retail company. The company has collected daily sales data for 2 years and wants you to identify the overall sales trend.
#   daily_sales = [220, 245, 210, 265, 230, 250, 260, 275, 240, 255, 235, 260, 245, 250, 225, 270, 265, 255, 250, 260]
#   ‚óè Explain how you would apply the Central Limit Theorem to estimate the average sales with a 95% confidence interval.
#   ‚óè Write the Python code to compute the mean sales and its confidence interval.

import numpy as np

daily_sales = [220, 245, 210, 265, 230, 250, 260, 275, 240, 255,
               235, 260, 245, 250, 225, 270, 265, 255, 250, 260]

# Mean and standard deviation
mean = np.mean(daily_sales)
std = np.std(daily_sales, ddof=1)
n = len(daily_sales)

# Standard error
SE = std / np.sqrt(n)

# 95% confidence interval
lower = mean - 1.96 * SE
upper = mean + 1.96 * SE

print(f"Mean Sales: {mean:.2f}")
print(f"95% Confidence Interval: ({lower:.2f}, {upper:.2f})")

```
